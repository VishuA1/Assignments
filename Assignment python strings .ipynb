### Question 2: Write a Python program to reverse a string without using [::-1]

- `sample input : `
    - `var = 'python'`
- `sample output :`
    - `result : nohtyp`

```python
# Please write the solution Here for above question 2:
def reverse(s):
    str = ""
    for i in s:
        str = i + str
    return str
s = "hello"
print("The original string is : ", end="")
print(s)
print("The reversed string After using loops is : ", end="")
print(reverse(s))

```

### Question 6: Write a Python program to remove all the vowels from a string.

- `sample input : `
    - `var = 'python'`
- `sample output :`
    - `result : pythn , because vowels are [AEIOUaeiou]`

```python
# Please write The solution Here for above question 6:
def rem_vowel(string):

    vowels = ['a','e','i','o','u']

    result = [letter for letter in string if letter.lower() not in vowels]

    result = ''.join(result)

    print(result)

string = "vishu -vishu is very good and loyal person"
rem_vowel(string)

```

### Question 7: Write a Python program to find the longest word in a string.

- `sample input : `
    - `var = 'python is Easy Language'`
- `sample output :`
    - `result : Language`

```python
# Please write The solution Here for above question 7:
def longestLength(words):

    res=max(words,key=len)

    print("The word with the longest length is:", res,

          " and length is ", len(res))

a = ["one", "two", "third", "four"]
longestLength(a)
```

### Question 8: Write a Python program to capitalize the first letter of each word in a string [dont use built-in-function title].

- `sample input : `
    - `var = 'python is easy'`
- `sample output :`
    - `result : Python Is Easy`

```python
# Please write The solution Here for above question 8:
import re
def convert_into_uppercase(a):
    return a.group(1) + a.group(2).upper()
s = "advance hpy b'day syamala"
result = re.sub("(^|\s)(\S)", convert_into_uppercase, s)
print(result)


```

### Question 10: write a python programme to find the sum of all the even characters based on Ascii values

- `sample input : `
    - `var = 'sharuk khan'`
- `sample output :`
    - `result : 464`


```python
import functools

test = "vishu"

print("The original list is : " + str(test))

res = [functools.reduce(lambda a, b: a + ord(b) , sub, 0) for sub in test]

print("Positional Summation List : " + str(res))
```
