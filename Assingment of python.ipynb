5.The program takes in an integer and prints the smallest divisor of the integer

```
n = int(input("Enter an integer: "))
smallest_divisor = None
for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
        smallest_divisor = i
        break
if smallest_divisor:
    print("The smallest divisor of", n, "is:", smallest_divisor)
else:
    print(n, "is a prime number")
```

6.Write a python program to check if it is a strong number or not

```
num =258
sum = 0
temp = num
while temp > 0:
  digit = temp % 10
  sum += digit ** 3
  temp //= 10
if num == sum:
  print(num,"is an strong number")
else:
  print(num,"is not an strong number")
```
